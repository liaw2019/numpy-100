#### 29. How to round away from zero a float array ? (★☆☆)

````python
import numpy as np

# Sample float array
Z = np.array([-1.5, -0.7, 0.0, 0.7, 1.5, 2.1, -2.9])
print("Original array:", Z)

# Round away from zero
# This works by taking the ceiling of the absolute value and then restoring the original sign.
rounded_away_from_zero = np.copysign(np.ceil(np.abs(Z)), Z)
print("Rounded away from zero:", rounded_away_from_zero)

# Example with positive numbers
Z_pos = np.array([1.2, 1.8, 0.1, 2.0])
rounded_pos = np.copysign(np.ceil(np.abs(Z_pos)), Z_pos)
print("\nOriginal positive array:", Z_pos)
print("Rounded positive away from zero:", rounded_pos)

# Example with negative numbers
Z_neg = np.array([-1.2, -1.8, -0.1, -2.0])
rounded_neg = np.copysign(np.ceil(np.abs(Z_neg)), Z_neg)
print("\nOriginal negative array:", Z_neg)
print("Rounded negative away from zero:", rounded_neg)
````

#### 35. How to compute ((A+B)*(-A/2)) in place (without copy)? (★★☆)

````python
A = np.ones(3)*1
B = np.ones(3)*2
np.add(A,B,out=B)
np.divide(A,2,out=A)
np.negative(A,out=A)
np.multiply(A,B,out=A)
print(B)
````

#### 36. Extract the integer part of a random array of positive numbers using 4 different methods (★★☆)

````python
import numpy as np

Z = np.array([1.2, 2.2, 2.6, 3.1, 5])
print("Original array:", Z)

# Method 1: Using astype(int)
print("Method 1 (astype(int)):", Z.astype(int))

# Method 2: Using np.trunc
print("Method 2 (np.trunc):", np.trunc(Z))

# Method 3: Using np.floor
print("Method 3 (np.floor):", np.floor(Z))

# Method 4: Subtracting the remainder (Z - Z % 1)
print("Method 4 (Z - Z % 1):", Z - Z % 1)
````
